# Other method to handle the missing values

```python
data = pd.read_csv("C:\\Users\\nikhi\\OneDrive\\Desktop\\Social Prachar\\Excel files\\Data Processing and Stats2.csv")
data
```

```python
# To check the count of missing values in each column
data.apply(lambda X: sum(X.isnull()),axis=0)
```

```python
# Filling the missing values- Age and Salary is continuous data type so we will place the missing values 
# Syntax Datasetname['Missing value column name'].fillna(mean)
data['Age'].fillna(data['Age'].mean(),inplace=True)
data['Salary'].fillna(data['Salary'].mean(),inplace=True)
data
```

```python
data['Gender'].fillna(data.Gender.mode()[0],inplace=True)
data
```
